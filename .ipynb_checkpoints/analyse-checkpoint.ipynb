```python
data = torch.load("./data_gen/FPT_NOTE_FRAMES.pt").contiguous().view(-1)
```

```python
notes_list = []
for i in data:
    if i <= 2:
        continue
    i = int(i)
    value = int(VOCABULARY["NOTES"][i].split("_")[-1])
    notes_list.append(value)
```

```python
from miditoolkit.midi import parser as mid_parser  
from miditoolkit.midi import containers as ct
import numpy as np
from miditoolkit.midi.containers import TimeSignature, KeySignature
```

```python
path_midi = 'midi_gen/test.mid'
```

```python
mido_obj = mid_parser.MidiFile()
beat_resol = mido_obj.ticks_per_beat
mido_obj.ticks_per_beat = 480
track = ct.Instrument(program=0, is_drum=False, name='example track')
mido_obj.instruments = [track]
mido_obj.time_signature_changes.append(TimeSignature(numerator=2, denominator=4, time=0))
mido_obj.key_signature_changes.append(KeySignature(key_name="C",time=0))
duration = 480
prev_end = 0
pitch = 60
for i in range(10):
    # create one note
    start = prev_end
    end = prev_end + duration
    pitch = pitch
    velocity = np.random.randint(1, 127)
    note = ct.Note(start=start, end=end, pitch=pitch, velocity=velocity)
    mido_obj.instruments[0].notes.append(note)
    
    # prepare next
    prev_end = end
    pitch += 1
print(mido_obj.time_signature_changes)
mido_obj.dump(path_midi)
```
